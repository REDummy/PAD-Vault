### 1. Random Intro
   ### a. Apa itu bilangan random: 
   Random number bukan berarti bilangan yang berbeda setiap waktu\
   Random artinya sesuatu yang tidak dapat diprediksi secara logika.
   ### b. Random number digenerate menggunakan algoritme yang disebut dengan pseudo random
   ### c. Generate Random Number  
   Menggunakan <span style="color:red">random</span> modul\
   Untuk memanggil random modul menggunakan **from numpy import random**
   ### d. Jenis data dan fungsi untuk random
   1. rand() --> float antara 0 sd 1
   2. rand(x) --> array 1D yang terdiri dari x random floats
   3. rand(x,y) --> array 2D yang terdiri dari x row dan setiap row terdiri dari 5 random floats
   4. randint(x) --> integer sd x tertentu
   5. randint(x,size=(y)) --> array 1D ukuran y dengan nilai rand sd nilai x
   6. randint(x,size=(y,z)) --> array 2D dengan y row dan setiap row terdiri dari z elemen, dengan nilai elemen adalah random integer dari 0 sd x.
   7. choice([]) --> mengenarte nilai random berdasarkan nilai array tertentu
   8. choice([],size=(x,y)) memilih dari array secara random dengan ukuran array yang dikembalikan adalah x,y

```python
#1. rand() --> float antara 0 sd 1
from numpy import random

x = random.rand()

print(x)
```

```python
#2. rand(x) --> array 1D yang terdiri x random float
x = random.rand(2)

print(x)
```

```python
#3. rand(x,y) --> array 2D yang terdiri dari x row dan setiap row terdiri dari 5 random floats
x = random.rand(2,3)

print(x)
```

```python
#4. randint(x) --> integer sd x tertentu
x = random.randint(9)

print(x)
```

```python
#5. randint(x,size=(y)) --> array 1D ukuran y dengan nilai rand sd nilai x
x = random.randint(9,size=(4))

print(x)
```

```python
#6. randint(x,size=(y,z)) --> array 2D dengan y row dan setiap row terdiri dari z elemen, 
# dengan nilai elemen adalah random integer dari 0 sd x.
x = random.randint(9,size=(4,2))

print(x)
```

```python
#7. choice() --> mengenerate nilai random berdasarkan nilai array tertentu,
# akan mengemballikan 1 nilai dari daftar list yang diberikan
   
from numpy import random

x = random.choice([3,5,7,9])

print(x)
```

```python
 ## 8. choice([],size=(x,y)) memilih dari array secara random dengan ukuran array yang dikembalikan adalah x,y
    
x = random.choice([3,5,7,9],size=(3,4))

print(x)
```

### 2. Random Data Distribution
1. **Apa itu Data Distribution** \
    Data distribution adalah list dari seluruh nilai yang mungkin dan seberapa sering nilai tersebut muncul\
    List yang terbentuk sangat penting ketika bekerja di bidang statistik dan data science\
    Modul random menawarkan metode yang mengembalikan distribusi data yang dihasilkan secara acak\z
    
2. **Random Distribution**\
    Random distribution merupakan himpunan bilangan acak yang mengikuti fungsi *probability density function*\
    **Probability Density Function**: A function that describes a continuous probability. i.e. probability of all values in an array.\
    Digenerate  menggunakan method <span style="color:red">choice() </span>  dari <span style="color:red">random</span> method.\
    Total dari probability number adalah 1

```python
#Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.
#The probability for the value to be 3 is set to be 0.1
#The probability for the value to be 5 is set to be 0.3
#The probability for the value to be 7 is set to be 0.6
#The probability for the value to be 9 is set to be 0

from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

```python
#Generate 2-D array with 3 rows, each containing 5 values.
from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3,5))

print(x)
```

### 3. Random Permutation
1. Random Permutation of Elements\
    Permutasi menunjuk pada melakukan penataan ulang (arrangement) dari element misal [3,2,1] adalah permutasi dari [1,2,3] dan sebaliknya\
    Menggunakan dua method yaitu : <span style="color:red">shuffle() </span>  dan <span style="color:red">permutation()</span> 


```python
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)

print(arr)
```

```python
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(random.permutation(arr))
```

```python
# 2. Normal Distribution
from numpy import random

y = random.normal(size=(2,3))
x = random.normal(size=(10))

print(y)
print(x)
```

```python
# Normal Distribution dengan paremeter
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
## Visualisasi Normal Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

```python
## 3. Binomial Distribution
from numpy import random

x = random.binomial(n=10, p=0.5, size=10)

print(x)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

```python
## Perbedaan normal distribution dan binomial
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

```python
## 4. Poisson Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.poisson(lam=2, size=10)

print(x)

sns.distplot(random.poisson(lam=2, size=1000), kde=False)

plt.show()
```

```python
## 5. Uniform Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.uniform(size=(2, 3))

print(x)

sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```

```python
## 6. Logistic Distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)

sns.distplot(random.logistic(size=1000), hist=False)

plt.show()
```

```python
## 7. Multinomial Distribution
from numpy import random

x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)

##Note: As they are generalization of binomial distribution 
##their visual representation and similarity of normal distribution is same as 
##that of multiple binomial distributions.
```

```python
## 8. Exponential Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.exponential(scale=2, size=(2, 3))

print(x)

sns.distplot(random.exponential(size=1000), hist=False)

plt.show()
```

```python
## 9. Chi-square Distribution
from numpy import random

x = random.chisquare(df=2, size=(2, 3))

print(x)

#--------------------
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.chisquare(df=1, size=1000), hist=False)

plt.show()
```

```python
## 10. Rayleigh Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.rayleigh(scale=2, size=(2, 3))

print(x)

sns.distplot(random.rayleigh(size=1000), hist=False)

plt.show()
```

```python
## 11. Pareto Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.pareto(a=2, size=(2, 3))

print(x)

sns.distplot(random.pareto(a=2, size=1000), kde=False)

plt.show()
```

```python
## 12. Zips Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=(2, 3))

print(x)

x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)

plt.show()
```
