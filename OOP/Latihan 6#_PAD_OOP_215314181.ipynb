### Savings Account

Write a program to maintain a savings account. 
The program should use a class named **SavingsAccount**  with instance variables for the customer's name and the account balance, and two methods named **makeDeposit** and **makeWithdrawal**. 
The makeWithdrawal method should deny withdrawals that exceed the balance in the account. See Fig. 7.22



```python
from savingsAccount import SavingsAccount

#Define Main Function
def main() :
    #input name and create object with name as parameter
    name = input("Enter person's name: ")
    account = SavingsAccount(name)

    #Always true loop
    while True:
        #Input action and 
        action = input("D = Deposit, W = Withdrawal, Q = Quit\nEnter D, W, or Q: ")
        
        #Match input with function
        if action.upper() == "D":
            amount = float(input("Enter amount to deposit: "))
            account.makeDeposit(amount)
        elif action.upper() == "W":
            amount = float(input("Enter amount to withdraw: "))
            account.makeWithdrawal(amount)
        elif action.upper() == "Q":
            print(f"End of transactions. Have a good day {account.name}.")
            break #Break out of loop
        else:
            print("Invalid action. Please try again.")


main()

```

## Cab Fare Management

Write a program for a cab owner to display the complete list of rides for his cabs.

He has both sedans and hatchbacks. The program should contain a class named **Cab** having two subclasses named **Sedan** and **Hatchback.** 

The Cab class should have instance variables for type of cab and number of kilometers driven. 

Each subclass should have a **caculateFare** method. 

The fare for sedans should be $2.0$ (2 dollar) per kilometer and the fare for hatchbacks should be $1.5 per kilometer.


After the data for all cabs has been entered, the program should display the total kilometers driven for both types of cabs. 

The program should also display the total kilometers driven for all cabs, and the total fare earned from all cabs. 

See Fig 7.23. The  program should use a list of objects.



```python
from cabs import Sedan, Hatchback
#Define main Function
def main() :
    cabList = []
    #Always true loop
    while True:
        #Input type and create object accordingly
        cabType = input("Enter cab type (Hatchback/Sedan): ")
        km = float(input("Enter the number of kilometers travelled: "))
        if cabType.lower() == "hatchback":
            cab = Hatchback(km)
        elif cabType.lower() == "sedan":
            cab = Sedan(km)
        else:
            print("Invalid cab type. Please try again.")
            continue
        #Add cab to list 
        cabList.append(cab)
        choice = input("Do you want to continue (Y/N)? ")
        if choice.upper() == "N":
            break #Break out of loop

    # display the list of cabs and their kilometers driven
    print("- Kilometers driven for each cab -")
    for cab in cabList:
        if isinstance(cab, Hatchback):
            print(f"HatchBack : {cab.km} kilometers")
        elif isinstance(cab,Sedan) :
            print(f"Sedan : {cab.km} kilometers")

    # calculate and display the total kilometers driven for each cab type
    hatchbackKm = sum([cab.km for cab in cabList if isinstance(cab, Hatchback)])
    sedanKm = sum([cab.km for cab in cabList if isinstance(cab,Sedan)])
    print(f"Total number of kilometers driven by Hatchbacks: {hatchbackKm}")
    print(f"Total number of kilometers driven by Sedans: {sedanKm}")

    # calculate and display the total fare earned from all cabs
    totalFare = sum([cab.calculateFare() for cab in cabList])
    print(f"Total fare earned from all cabs (in dollars): {totalFare:.1f}")

main()
```
