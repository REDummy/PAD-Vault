### DataFrame
     1. Berukuran 2 dimensi: size-mutable, berpotensi untuk berbagai struktur tabular 2 dimensi (rows and column).
     2. Terdiri dari 3 prinsipal component : data, rows , column 
     


```python
## Membuat DataFrame 
import pandas as pd

## Constructor
df=pd.DataFrame()
print(df)

## List of string
lst=['Informatika','Universitas','Sanata','Dharma']

## Calling dataframe constructor ke list
df=pd.DataFrame(lst)
print(df)
```
