### loc
### .loc['nama baris','nama kolom']
iloc harus mengetahui indexnya, sedangkan jika tidak diketahui indexnya gunakan loc

```python
import pandas as pd
databaru=pd.read_csv('kapal_titanic.csv',index_col='embarked')
```

```python
# ambil column 'age' dari embarked='S'
# perhatikan hasilnya apakah berupa series atau dataframe
databaru.loc['S']
```

```python
# tampilkan embarked='S' dengan column age bertipe series
databaru.loc['S',['age']]
```
